```python
import pandas as pd
import numpy as np
from functools import reduce
```

```python
# reminder: on a new session, for this to work, run scraper.ipynb first
%store -r zip_codes
# ['21201', '21202', '21203', '21205', '21206', '21209', '21210', 
# '21211', '21212', '21213', '21214', '21215', '21216', '21217', 
# '21218', '21223', '21224', '21225', '21229', '21230', '21231', 
# '21233', '21239', '21270', '21279', '21281', '21297']
```

```python
def filter_and_clean_df(df, columns_to_drop):
  filtered_df = df[(df['geo_level'] == 'zcta') & (df['geo_name'].isin(zip_codes)) 
                # | (df['geo_level'] == 'county') & (df['geo_name'].str.contains('Baltimore city'))
              ]

  if filtered_df.empty:
    filtered_df = df[(df['geo_level'] == 'county') & (df['geo_name'].str.contains('Baltimore city'))]

  cleaned_df = filtered_df.drop(columns=columns_to_drop)

  cleaned_df.dropna(inplace=True)

  cleaned_df.reset_index(drop=True, inplace=True)

  return cleaned_df
```

```python
df = pd.read_csv('..\\bwdc_datasets\\health_insurance_coverage.csv')

df
```

```python
columns_to_drop = ['geo_level', 'year', 'state_name', 'pct_with_health_insurance_white_alone', 'pct_with_health_insurance_asian_alone', 'pct_with_health_insurance_hispanic', 'quartiles_pct_with_health_insurance']

health_insurance_coverage = filter_and_clean_df(df, columns_to_drop)

health_insurance_coverage
```

```python
df = pd.read_csv('..\\bwdc_datasets\\number_of_loan_apps.csv')

df.head()
```

```python
columns_to_drop = ['state_name']

number_of_loan_apps = filter_and_clean_df(df, columns_to_drop)

number_of_loan_apps
```

```python
all_loans_avg_amount = number_of_loan_apps[number_of_loan_apps['loan_type'] == 'All']['loan_amount'].mean()
loans_7a_avg_amount = number_of_loan_apps[number_of_loan_apps['loan_type'] == '7(a) Loans']['loan_amount'].mean()
loans_504_avg_amount = number_of_loan_apps[number_of_loan_apps['loan_type'] == '504 Loans']['loan_amount'].mean()

all_loans_avg_number = number_of_loan_apps[number_of_loan_apps['loan_type'] == 'All']['number_of_loans'].mean()
loans_7a_avg_number = number_of_loan_apps[number_of_loan_apps['loan_type'] == '7(a) Loans']['number_of_loans'].mean()
loans_504_avg_number = number_of_loan_apps[number_of_loan_apps['loan_type'] == '504 Loans']['number_of_loans'].mean()

# create new df with averages
averages_df = pd.DataFrame({
    'loan_type': ['All', '7(a) Loans', '504 Loans'],
    'average_loan_amount': [all_loans_avg_amount, loans_7a_avg_amount, loans_504_avg_amount],
    'average_number_of_loans': [all_loans_avg_number, loans_7a_avg_number, loans_504_avg_number]
})

averages_df
```

```python
# list of DataFrames to merge
dfs = [health_insurance_coverage, 
    # high_speed_broadband_internet_usage, 
    home_ownership_rate, 
    # number_of_loan_apps, 
    percent_broadband_internet_subscription_by_households, rate_of_bachelors_or_higher]

# merge dfs
merged_df = reduce(lambda left, right: pd.merge(left, right, on='geo_name', how='inner'), dfs)

merged_df
```
